## 1. Two Sum
Trouver deux indices i et j tels que nums[i] + nums[j] = target.


```python
def two_sum(nums, target):
    seen  = {} #dictionary to store number and its index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen :
            return [seen [complement], i]
        seen [num] = i
    return []
```

# 2. Valid Anagram
Vérifier si deux chaînes sont des anagrammes.
Verifier d'abord si la chaines est alphanumérique 

```python
from collections import Counter
def is_anagram(s1, s2):
    s1 = ''.join(filter(str.isalnum, s1)).lower()
    s2 = ''.join(filter(str.isalnum, s2)).lower()
    return Counter(s1) == Counter(s2)
#counter crée un dictionnaire avec le nombre d'occurrences de chaque caractère
#exemple: Counter("anagram") -> {'a': 3, 'n': 1, 'g': 1, 'r': 1, 'm': 1}
```

# 3. Valid Palindrome
A phrase is a palindrome if, after converting all uppercase letters into lowercase letters and removing all non-alphanumeric characters, it reads the same forward and backward. Alphanumeric characters include letters and numbers

```python
def is_palindrome(s):
    filtered_chars = [char.lower() for char in s if char.isalnum()]
    return filtered_chars == filtered_chars[::-1]
```

# 4. isomorphic Strings
Two strings s and t are isomorphic if the characters in s can be replaced to get t
exemple: "egg" and "add" are isomorphic because 'e' -> 'a' and 'g' -> 'd'

```python
#zip donne une liste de tuples
def isIsomorphic(s, t):
    map_s_t = {}
    map_t_s = {}
    for cs, ct in zip(s, t): 
        if (cs in map_s_t and map_s_t[cs] != ct) or \
           (ct in map_t_s and map_t_s[ct] != cs): 
        #si le car de s est déjà mappé à un autre caractère t ou vice versa
            return False
        map_s_t[cs] = ct
        map_t_s[ct] = cs
    return True
```

# 5. First Unique Character in a String
Trouver le premier caractère unique dans une chaîne.

```python
def firstUniqChar(s):
    from collections import Counter
    freq = Counter(s)
    for i, c in enumerate(s):
        if freq[c] == 1:
            return i
    return -1
```

# 6. Longest Substring Without Repeating Characters
Trouver la longueur de la plus longue sous-chaîne sans caractères répétés.
s = "abcabcbb" → résultat 3 ("abc")
 Méthode optimale : Sliding Window (O(n))
Utiliser deux pointeurs left et right.

Maintenir une fenêtre contenant uniquement des caractères uniques.

Utiliser un dictionnaire pour enregistrer le dernier index de chaque caractère.

```python
def lengthOfLongestSubstring(s):
    char_index_map = {}
    left = 0
    max_length = 0

    for right, char in enumerate(s):
        if char in char_index_map and char_index_map[char] >= left:
            left = char_index_map[char] + 1
        char_index_map[char] = right
        max_length = max(max_length, right - left + 1)

    return max_length

#enumerate donne l'index et la valeur en même temps
```
